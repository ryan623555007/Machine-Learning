```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
```

```python
train = pd.read_csv("./train.csv")
test  = pd.read_csv("./test.csv")
print ('训练数据集:',train.shape,'测试数据集:',test.shape)
```

```python
rowNum_train=train.shape[0]
rowNum_test=test.shape[0]
print(rowNum_train,rowNum_test)
```

```python
full=train.append(test,ignore_index=True)
print(full.shape)
```

```python
full.info()
full['Age']=full['Age'].fillna( full['Age'].mean() )
full['Fare']=full['Fare'].fillna( full['Age'].mean() )
full.info()
```

```python
full.head()
```

```python
full['Embarked'].head()
```

```python
full['Embarked'].value_counts()
```

```python
full['Embarked']=full['Embarked'].fillna( 'S' )
```

```python
full['Cabin'].head()
```

```python
full['Cabin']=full['Embarked'].fillna( 'N' )
```

```python
full.head()
```

```python
full.info()
```

特征提取

```python
full['Sex'].head()
```

```python
sex_mapDict={'male':1,'female':2}
full['Sex']=full['Sex'].map(sex_mapDict)
```

```python
full.head()
```

```python
full['Embarked'].head()
```

```python
embarkedDf=pd.DataFrame()
embarkedDf=pd.get_dummies( full['Embarked'] , prefix='Embarked' )
embarkedDf.head()
```

```python
full=pd.concat([full,embarkedDf],axis=1)
full.drop('Embarked',axis=1,inplace=True)
full.head()
```

```python
pclassDf=pd.DataFrame()
pclassDf=pd.get_dummies( full['Pclass'],prefix='Pclass' )
pclassDf.head()
```

```python
full=pd.concat([full,pclassDf],axis=1)
full.drop('Pclass',axis=1,inplace=True)
full.head()
```

```python
full['Name'].head()
```

```python
def getTitle(name):
    str1=name.split( ',' )[1]
    str2=str1.split( '.' )[0]
    str3=str2.strip()
    return str3
```

```python
titleDf=pd.DataFrame()
titleDf['Title']=full['Name'].map(getTitle)
titleDf.head()
```

```python
title_mapDict = {
                    "Capt":       "Officer",
                    "Col":        "Officer",
                    "Major":      "Officer",
                    "Jonkheer":   "Royalty",
                    "Don":        "Royalty",
                    "Sir" :       "Royalty",
                    "Dr":         "Officer",
                    "Rev":        "Officer",
                    "the Countess":"Royalty",
                    "Dona":       "Royalty",
                    "Mme":        "Mrs",
                    "Mlle":       "Miss",
                    "Ms":         "Mrs",
                    "Mr" :        "Mr",
                    "Mrs" :       "Mrs",
                    "Miss" :      "Miss",
                    "Master" :    "Master",
                    "Lady" :      "Royalty"
                    }
titleDf['Title'] = titleDf['Title'].map(title_mapDict)
titleDf=pd.get_dummies(titleDf['Title'])
titleDf.head()
```

```python
full=pd.concat([full,titleDf],axis=1)
full.drop('Name',axis=1,inplace=True)
full.head()
```

```python
sum = lambda a,b: a + b
print ("相加后的值为 : ", sum(10,20))
```

```python
cabinDf=pd.DataFrame()
full['Cabin']=full['Cabin'].map( lambda c:c[0] )
cabinDf=pd.get_dummies( full['Cabin'],prefix='Cabin' )
cabinDf.head()
```

```python
full=pd.concat([full,cabinDf],axis=1)
full.drop('Cabin',axis=1,inplace=True )
full.head()
```

```python
familyDf=pd.DataFrame()
familyDf['FamilySize']=full['Parch']+full['SibSp']+1
familyDf['Family_Single']=familyDf['FamilySize'].map(lambda s:1 if s==1 else 0 )
familyDf['Family_Small']=familyDf['FamilySize'].map(lambda s:1 if 1<s<=4 else 0 )
familyDf['Family_Large']=familyDf['FamilySize'].map(lambda s:1 if s>4 else 0 )
familyDf.head()
```

```python
full=pd.concat([full,familyDf],axis=1)
full.head()
```

```python
full.shape
```

```python
corrDf=full.corr()
corrDf
```

```python
corrDf['Survived'].sort_values(ascending=False)
```

```python
full_X=pd.concat([titleDf,
                 pclassDf,
                 familyDf,
                 full['Fare'],
                 cabinDf,
                 embarkedDf,
                 full['Sex']
                 ],axis=1)
full_X.head()
```

构建模型

```python
sourceRow=891
source_X = full_X.loc[0:sourceRow-1,:]
source_y = full.loc[0:sourceRow-1,'Survived']   
pred_X = full_X.loc[sourceRow:,:]
```

```python
from sklearn.cross_validation import train_test_split 
train_X, test_X, train_y, test_y = train_test_split(source_X ,
                                                    source_y,
                                                    train_size=.8)
print ('原始数据集特征：',source_X.shape, 
       '训练数据集特征：',train_X.shape ,
      '测试数据集特征：',test_X.shape)

print ('原始数据集标签：',source_y.shape, 
       '训练数据集标签：',train_y.shape ,
      '测试数据集标签：',test_y.shape)
```

```python
source_y.head()
```

```python
source_X.head()
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit( train_X , train_y )
model.score(test_X , test_y )
```

```python
pred_Y = model.predict(pred_X)
passenger_id = full.loc[sourceRow:,'PassengerId']
predDf = pd.DataFrame( 
    { 'PassengerId': passenger_id , 
     'Survived': pred_Y } )
predDf.shape
predDf.head()
predDf.to_csv( 'titanic_pred.csv' , index = False )
```
